## Outliers


### Finding outliers


Find out if outliers exist in the column `ConvertedComp` using a box plot?


```python
# your code goes here
df[['ConvertedComp']].plot(kind='box', figsize=(8,6))
plt.show()
```

Find out the Inter Quartile Range for the column `ConvertedComp`.


```python
len(df['ConvertedComp'].dropna())
```

```python
data = np.array(df['ConvertedComp'].dropna())
```

```python
# your code goes here
q1, q3 = np.percentile(data, [25, 75])
print(q3, q1)
```

```python
IQR = q3-q1
IQR
```

Find out the upper and lower bounds.


```python
# your code goes here
upper = q3 + 1.5*IQR
lower = q1 - 1.5*IQR
lower, upper
```

Identify how many outliers are there in the `ConvertedComp` column.


```python
# your code goes here
df[df['ConvertedComp'] < upper]['Respondent'].count()
```

Create a new dataframe by removing the outliers from the `ConvertedComp` column.


```python
# your code goes here
df_new = df[df['ConvertedComp'] < upper].reset_index(drop=True)
df_new.tail()
```
